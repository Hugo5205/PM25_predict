```python
import pymongo
import pandas as pd
#from bson.objectid import ObjectId #這東西再透過ObjectID去尋找的時候會用到
from sklearn.externals import joblib #jbolib模块

if __name__ == "__main__":
    
    # 與 MongoDB連線
    client = pymongo.MongoClient(host="mongodb", port=27017)
    # 指定為 test 資料庫
    db = client.hugo
    # 指定 temp_humidity 集合, MongoDB的每個資料庫又包含許多集合(collection), 類似於關聯性資料庫中的表
    collection = db.pm25_testmodel
    
    cursor = collection.find({},{"_id":0}) #此處須注意，其回傳的並不是資料本身，你必須在迴圈中逐一讀出來的過程中，它才真的會去資料庫把資料撈出來給你。
    data = [d for d in cursor] #這樣才能真正從資料庫把資料庫撈到python的暫存記憶體中。
    df = pd.DataFrame(data,columns=['timestamp','Temperature','Humidity','PM25'])
    df["hour"] = df['timestamp'].astype(str).str.split(' ',expand = True)[1].str.split(':',expand = True)[0].astype(int) + 1
    #df["hour"] = (!(df["hour"] == 24)).astype('int')
df
```

```python
for i in range(len(df)):
    if df["hour"][i] == 24:
        df["hour"][i] = 0
X_test = df[['hour', 'Temperature', 'Humidity']]
y_test = df['PM25']
df
```

```python
from sklearn.externals import joblib
import numpy as np
true = np.array(y_test).astype(float).astype(int)
#读取Model
#分類決策樹
DecisionTreeClassifier = joblib.load('./DecisionTreeClassifier.pkl')
DecisionTreeClassifier_pred = DecisionTreeClassifier.predict(X_test)
#線性回歸
LinearRegression = joblib.load('./LinearRegression.pkl')
LinearRegression_pred = LinearRegression.predict(X_test)
#回歸決策樹
DecisionTreeRegressor = joblib.load('./DecisionTreeRegressor.pkl')
DecisionTreeRegressor_pred = DecisionTreeRegressor.predict(X_test)
#Elastic_Net
Elastic_Net = joblib.load('./Elastic_Net.pkl')
Elastic_Net_pred = Elastic_Net.predict(X_test)
#隨機森林
RandomForestClassifier = joblib.load('./RandomForestClassifier.pkl')
RandomForestClassifier_pred = RandomForestClassifier.predict(X_test)
#SVR(高斯)
SVR_Gaussian_Kernel = joblib.load('./SVR_Gaussian_Kernel.pkl')
SVR_Gaussian_Kernel_pred = SVR_Gaussian_Kernel.predict(X_test)
#测试读取后的Model
```

```python
#模型評分
import numpy as np
from statistics import mean 

true = np.array(y_test).astype(float)

def mean_absolute_percentage_error(y_true, y_pred): 
    #y_true, y_pred = check_arrays(y_true, y_pred)
    #y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100

print("分類決策樹 :",mean_absolute_percentage_error(true, DecisionTreeClassifier_pred))
print("線性回歸 :",mean_absolute_percentage_error(true, LinearRegression_pred))
print("回歸決策樹 :",mean_absolute_percentage_error(true, DecisionTreeRegressor_pred))
print("Elastic_Net :",mean_absolute_percentage_error(true, Elastic_Net_pred))
print("隨機森林 :",mean_absolute_percentage_error(true, RandomForestClassifier_pred))
print("SVR(高斯) :",mean_absolute_percentage_error(true, SVR_Gaussian_Kernel_pred))
```
